```python
import pandas as pd

data = pd.read_csv("sales.csv")
data
```

```python
import matplotlib.pyplot as plt

plt.bar(data["Product"], data["Sales"])
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Sales by Product")
plt.show()
```

```python
plt.plot(data["Date"], data["Sales"], marker="o")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Sales Trend Over Time")
plt.show()
```

```python
category_sales = data.groupby("Category")["Sales"].sum()

plt.pie(category_sales, labels=category_sales.index, autopct="%1.1f%%")
plt.title("Sales Distribution by Category")
plt.show()
```

```python
import seaborn as sns

pivot_table = data.pivot_table(
    values="Sales",
    index="Category",
    columns="Product",
    aggfunc="sum"
)

sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Sales Heatmap by Category and Product")
plt.show()
```

```python


1. Electronics category contributes the highest sales, accounting for nearly 73% of total revenue.
2. Laptop is the top-performing product with the highest individual sales.
3. Sales peaked on 2nd January and dropped sharply on 3rd January, indicating fluctuating daily demand.
4. Clothing category shows moderate sales contribution compared to Electronics.
5. Accessories contribute the least to overall sales.
```
